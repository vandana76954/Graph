```python
df= pd.read_csv("https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv")
```

```python
df.plot(kind='hist')
#histogram with tell us value and frequency range of it.
#It will tell us behaviour  of  dataset
```

```python
#lets take one column to understand histogram
df["SepalLengthCm"].plot(kind="hist")
#here x-axis represents value range and y-axis represents frequency of it
#it represents how many sepal length are in between 4.5 to 4.7 which is 8 dataset values.
#how many datsets values are in between 4.7 to 5.0 which maybe 23.
```

```python
#In order to rotate the histogram graph
df["SepalLengthCm"].plot(kind="hist",orientation="horizontal")
```

```python
df
```

```python
#hist() will craete histogram of all the columns
df.hist(figsize=(20,10),color='g',alpha=.5) 
#meaning of alpha is intensity of color alpha range is between 0 to 1 like 0.9 will be high  0.2 to will be low
```

```python
#Boxplot
df.plot(kind='box',figsize=(20,10))
```

```python
df.plot(kind='box',figsize=(20,10),color={"boxes" : "g","whiskers" : "r"},vert=False)
```

```python
#Area graph which is similar to line graph
df.plot(kind="area",figsize=(20,10),alpha=0.7,stacked=False)
#here stacked=False.It is not going to overlap
```

```python
df.plot(kind="area",figsize=(20,10),alpha=0.7,stacked=True)
#here stacked=True.It is going to overlap
```

```python
#Scatter plot is used to see relationship between two variables
df.plot.scatter(x="SepalLengthCm",y="PetalLengthCm")
#how Sepal  length is increasing w.r.t Petal length .
#we can say sepal length is increasing when petal length is increasing
#here we can say it has stroong relatioship
```

```python
#if we want to control size of circle w.r.t to other variable
df.plot.scatter(x="SepalLengthCm",y="PetalLengthCm",c="SepalWidthCm" ,s= df["PetalWidthCm"]*60)
```

```python
df2=sns.load_dataset("tips")
df2
```

```python
sns.relplot(x="total_bill",y="tip",data=df2)
```

```python
#Based on this restaurant can taken decision which area to be expanded
df2["smoker"].value_counts()
```

```python
#Based on smoker lets plot  scatterplot
sns.relplot(x="total_bill",y="tip",data=df2,hue="smoker")
#here hue will plot graph by grouping data on the basis of non smokers and smokers.
```

```python
sns.relplot(x="total_bill",y="tip",data=df2,style="smoker")
#style will group two classes as per two styles
```

```python
sns.relplot(x="total_bill",y="tip",data=df2,style="size",)
```

```python
#A person who has paid highest bill came in group(size) of 3
#A person who has paid lowest bill came in group(size) of  1 and 2
```

```python
#Lets use hue and style together
sns.relplot(x="total_bill",y="tip",data=df2,style="time",hue="smoker")
#Person who paid highest bill was smoker and he came at dinner time
#Person who paid loowest bill was smoker and he came at dinner time
```

```python
sns.relplot(x="total_bill",y="tip",data=df2,size="size")
#this is also performing groupby only
```

```python
sns.relplot(x="total_bill",y="tip",kind="line",data=df2,)
```

```python
#lets create two seperate graph for the time columns with two different classes lunch and Dinner
sns.relplot(x="total_bill",y="tip",col="time",data=df2)
```

```python
sns.relplot(x="total_bill",y="tip",col="day",data=df2)
#On which day people people are paying more billls on sat and sun
```

```python
sns.relplot(x="total_bill",y="tip",col="sex",data=df2)
#MAle are paying more tips and bill
```

```python
#for  categorical variable like day
sns.catplot(x="day",y="total_bill",data=df2)
#On sat sunday bill was highest
```

```python
sns.catplot(x="day",y="tip",data=df2)
#On dat tip got was highest
#least on friday
```

```python
sns.catplot(x="day",y="tip",data=df2,kind="swarm")
```

```python
#boxplot
sns.catplot(x="day",y="tip",data=df2,kind="box")
```

```python
sns.catplot(x="day",y="tip",data=df2,kind="box",hue="sex")
```
